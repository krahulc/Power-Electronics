```python
from __future__ import division
import numpy as np

from scipy.integrate import ode
import math

import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.pyplot as plt
x0, t0 = [1, 0]

def dx_dt(t, x, *arg1):
    return math.sin(t)

rahul = ode(dx_dt)
rahul.set_initial_value(x0, t0)
rahul.set_integrator("dopri5")

x = []
y = []

t_end = 30
dt = 0.1

while rahul.successful() and rahul.t < t_end:
        y.append(rahul.t + dt)
        x.append(rahul.integrate(rahul.t + dt))

plt.plot(y, x)
```

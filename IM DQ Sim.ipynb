```python
import numpy as np
from scipy import linalg
import matplotlib.pylab as plt
import math
from scipy import integrate
from scipy.integrate import ode
from scipy.integrate import quad
from scipy.integrate import trapz
import seaborn as sns
%matplotlib inline
```

```python
Vm = 105*1.414    #Phase peak voltage                 
freq = 400
w=2*np.pi *freq    
J = 0.000675        #moment of inertia
T_l = 1.1          #load torque
B= 0.00006          #coefficient of friction 
P=4               #no of poles
Rs=.047 #stator resistance
Rr=.05 #rotor resistance
Lm=8.71/w #mutual inductance
Lls=.433/w #stator leakage inductance
Llr=.623/w #rootor leakage inductance
Ls=Lm + Lls   #stator inductance
Lr=Lm + Llr   # rotor inductance
```

```python
R=([Rs,0,0,0],[0,Rs,0,0],[0,0,Rr,0],[0,0,0,Rr])      
L=([Ls,0,Lm,0],[0,Ls,0,Lm],[Lm,0,Lr,0],[0,Lm,0,Lr])
G=([0,0,0,0],[0,0,0,0],[0,Lm,0,Lr],[-Lm,0,-Lr,0])

invL=np.linalg.inv(L)

vdq0 = vdq01 = vdq02 = vdq03 = (np.zeros((2,1)) for i in range(4))
v0, v1, v2, v3, i, k1, k2, k3, k4 = (np.zeros((4,1)) for i in range(9))
i_abc=np.zeros((3,1))

h=.0001
n=400/freq
step = int(n/h)

wr, Te, t, ias, ibs, ics  = ([0]*(step+1) for i in range(6)) #initialise all empty arrays with zeroes

def dq(k,h):
    c = ([1,-.5,-.5],[0,3**0.5/2,-3**0.5/2])
    return ((2./3)*Vm)*np.dot(c,([np.sin(w*(t[k]+h))] ,[np.sin(w*(t[k]+h) -(2./3)*np.pi)] ,[np.sin(w*(t[k]+h) + (2./3)*np.pi)]))

for k in range(0,step ,1):  
    
    vdq0 = dq(k,0) 
    vdq1 = dq(k,h/2) 
    vdq2 = dq(k,h/2) 
    vdq3 = dq(k,h) 
    
    v0[0], v0[1] = vdq0[0], vdq0[1]
    v1[0], v1[1] = vdq1[0], vdq1[1]
    v2[0], v2[1] = vdq2[0], vdq2[1]
    v3[0], v3[1] = vdq3[0], vdq3[1]
         
    k1=np.dot(invL ,(v0-(np.dot(R,i) + wr[k]*np.dot(G,i))))
    k2=np.dot(invL ,((v1+ (h/2)*k1)- (np.dot(R,(i+(h/2)*k1)) + wr[k]*np.dot(G,(i+ (h/2)*k1)))))
    k3=np.dot(invL ,((v2+ (h/2)*k2)- (np.dot(R,(i+(h/2)*k2)) + wr[k]*np.dot(G,(i+ (h/2)*k2)))))                                                                                   
    k4=np.dot(invL ,((v3+ h*k3)-(np.dot(R,(i+ h*k3))+ wr[k]*np.dot(G,(i+ h*k3)))))
    
    i= i + (h/6)*(k1+2*k2+2*k3+k4)
    ias[k+1]=i[0]
    ibs[k+1]=(-.5)*i[0]+(.866)*i[1]
    ics[k+1]=(-.5)*i[0]-(.866)*i[1]
    Te[k+1] = (3*Lm*((i[2] * i[1]) -(i[3] *i[0])))
    wr[k+1] = (Te[k] - T_l )*P*h/(2*J) + (1-(B*h/J))*wr[k] 
    t[k+1] = t[k] + h
```

```python
plt.plot(t,Te)
plt.show()

wr = np.dot(60/(4*math.pi),wr) #in RPM
plt.plot(t,wr)
plt.show()

plt.plot(t,ias)
plt.plot(t,ibs)
plt.plot(t,ics)
plt.axis([0.4,0.55,-300,300])
```
